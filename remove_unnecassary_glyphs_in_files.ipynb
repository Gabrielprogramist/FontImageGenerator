```python
russian_chars = list(range(0x0410, 0x0450))+[0x0401, 0x0451]  # А-я
kazakh_chars = [0x04D8, 0x04D9, 0x0492, 0x0493, 0x049A, 0x049B, 0x04A2, 0x04A3, 0x04E8, 0x04E9, 0x04B0, 0x04B1, 0x04AE, 0x04AF, 0x04BA, 0x04BB, 0x0406, 0x0456]
kazakh_cyrillic = [chr(i) for i in russian_chars+kazakh_chars]
kazakh_cyrillic
```

```python
import os
import shutil

def check_objects_in_folder(folder_path, required_objects):
    # Check if the required objects are present in the folder
    objects_in_folder = set()
    files = os.listdir(folder_path)
    
    # Check each file in the folder
    for file_name in files:
        # If the file is an image, extract the object name from the file name
        if file_name.endswith('.jpg') or file_name.endswith('.png'):
            object_name = file_name.split('_')[1]
            objects_in_folder.add(object_name)
    
    # Check if all required objects are in the folder
    if set(required_objects).issubset(objects_in_folder):
        # Keep only files that correspond to the required objects
        for file_name in files:
            object_name = file_name.split('_')[1]
            if object_name not in required_objects:
                os.remove(os.path.join(folder_path, file_name))
    else:
        # Remove the entire folder if one or more objects are missing
        shutil.rmtree(folder_path)

# Path to the folder containing subfolders with images
all_files_path = "all_available_glyphs"

# List of objects that must be present in each folder
required_objects = kazakh_cyrillic

# Check each subfolder in the main folder
for folder_name in os.listdir(all_files_path):
    folder_path = os.path.join(all_files_path, folder_name)
    if os.path.isdir(folder_path):
        check_objects_in_folder(folder_path, required_objects)
```
